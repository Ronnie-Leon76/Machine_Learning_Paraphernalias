```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
# from sklearn.datasets import load_diabetes
```

```python
my_data = pd.read_csv('./Bike-Sharing-Dataset/day.csv',usecols=['season','holiday','weekday','weathersit','cnt'])
my_data.head()
```

```python
my_data.tail()
```

```python
sns.relplot(
    x='season', y='weekday', hue='cnt', data=my_data)
plt.show()
```

```python
sns.relplot(
    x='season', y='weathersit', hue='cnt', data=my_data)
plt.show()
```

```python
my_data.sample(frac=1, replace= True).head()
#sample will allow us to get a random rows by using frac and replace parameter
```

```python
my_data.shape
```

```python
my_data.info()
```

```python
features = my_data.iloc[:,:-1]
features.head()
```

```python
target = my_data.iloc[:,-1]
target.head()
```

```python
from sklearn.model_selection import train_test_split
training_features, test_features, training_target, test_target = train_test_split(features, target,test_size=0.1, random_state=0)
```

```python
training_features.shape
```

```python
training_target.shape
```

```python
test_features.shape
```

```python
test_target.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
my_regressor = DecisionTreeRegressor(criterion='mse', min_samples_leaf=6)
my_regressor
```

```python
my_regressor.fit(training_features, training_target)
```

```python
prediction = my_regressor.predict(test_features)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print('The mean square error is:', mean_squared_error(test_target, prediction))
print('The root mean square error is:', np.sqrt(mean_squared_error(test_target, prediction)))
print('The mean absolute error is:', mean_absolute_error(test_target, prediction))
print('The r-square:',r2_score(test_target, prediction))
```

```python
# Plot outputs
plt.scatter(test_target, prediction,color='black')
# plt.plot(test_target, prediction, color='blue', linewidth=3)

# plt.xticks(())
# plt.yticks(())

plt.show()
```
