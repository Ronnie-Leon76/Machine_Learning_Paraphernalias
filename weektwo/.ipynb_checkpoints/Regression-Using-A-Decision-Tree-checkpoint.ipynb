```python
data = pd.read_csv('./Dataset/day.csv', usecols=['season','holiday','weekday','workingday','cnt'])
data.head(100)
```

```python
features = data.iloc[: , 0:4]
features.head()
```

We are going to split data by columns input, output

```python
features = data.iloc[: ,  :-1]
features.head()
```

```python
target = data.iloc[: ,-1: ]
target.head()
```

```python
target = data.iloc[: ,-1: ]
target.head()
```

```python
from sklearn.model_selection import train_test_split
training_features, testing_features, training_target, testing_target = train_test_split(features, target, test_size=0.3,random_state = 0)
```

```python
from sklearn.model_selection import train_test_split
training_features, testing_features, training_target, testing_target = train_test_split(features, target, test_size=0.3,
                                                                                        random_state = 0 )
```

```python
print(training_features.shape, training_target.shape, '\n', testing_features.shape, testing_target.shape)
```

```python
print(training_features.shape, training_target.shape, '\n',testing_features.shape, testing_target.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(criterion='mse', min_samples_leaf=5)
```

```python
dtr.fit(training_features, training_target)
```

```python
model_pred = dtr.predict(testing_features)
model_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
mse = mean_squared_error(testing_target, model_pred)
mae = mean_absolute_error(testing_target, model_pred)
rmse =np.sqrt(mean_squared_error(testing_target, model_pred))
r_square = r2_score(testing_target, model_pred)
print(f'mean square error:{mse}\nmean absolute error:{mae}\nroot mean square error:{rmse}\nr_square: {r_square}')
```

```python
#evaluating loss functions
g=plt.scatter(testing_target, model_pred )
g.axes.set_yscale('log')
g.axes.set_xscale('log')
g.axes.set_xlabel('True Values ')
g.axes.set_ylabel('Predictions ')
g.axes.axis('equal')
g.axes.axis('square')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(testing_target, model_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(model_pred), max(testing_target))
p2 = min(min(model_pred), min(testing_target))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```
